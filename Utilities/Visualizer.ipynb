# SETUP

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import pickle
sns.set(rc={'figure.figsize': (16, 9)})
```

```python
algorithm = 'OLD_DDQN'
environment = 'TicTacToe'
representation = 'Graphic'
opponent = 'MinMaxRandom'
```

```python
agent_turn = 'Random'

config_name = algorithm + '_' + environment + '_' + representation + '_' + opponent + '_' + agent_turn
data_path = '../Results/' + config_name + '/scores.pkl'
f = open(data_path, 'rb')
results_random = pickle.load(f)
f.close()
data_random_reward = pd.DataFrame(np.array([(key, value[i][0]) for key, value in results_random.items() for i in
                 range(len(value))]), columns=['episode', 'reward'])
data_random_reward['rolling'] = data_random_reward.reward.rolling(100).mean()
data_random_length = pd.DataFrame(np.array([(key, value[i][1]) for key, value in results_random.items() for i in
                 range(len(value))]), columns=['episode', 'length'])
data_random_length['rolling'] = data_random_length.length.rolling(100).mean()
```

```python
agent_turn = 'First'

config_name = algorithm + '_' + environment + '_' + representation + '_' + opponent + '_' + agent_turn
data_path = '../Results/' + config_name + '/scores.pkl'
f = open(data_path, 'rb')
results_first = pickle.load(f)
f.close()
data_first_reward = pd.DataFrame(np.array([(key, value[i][0]) for key, value in results_first.items() for i in
                 range(len(value))]), columns=['episode', 'reward'])
data_first_length = pd.DataFrame(np.array([(key, value[i][1]) for key, value in results_first.items() for i in
                 range(len(value))]), columns=['episode', 'length'])
```

```python
agent_turn = 'Second'

config_name = algorithm + '_' + environment + '_' + representation + '_' + opponent + '_' + agent_turn
data_path = '../Results/' + config_name + '/scores.pkl'
f = open(data_path, 'rb')
results_second = pickle.load(f)
f.close()
data_second_reward = pd.DataFrame(np.array([(key, value[i][0]) for key, value in results_second.items() for i in
                 range(len(value))]), columns=['episode', 'reward'])
data_second_length = pd.DataFrame(np.array([(key, value[i][1]) for key, value in results_second.items() for i in
                 range(len(value))]), columns=['episode', 'length'])
```

# EPISODE REWARD PLOT

```python
fig, ax = plt.subplots()
ax = sns.lineplot(x='episode', y='reward', data=data_random_reward, label='random')
ax = sns.lineplot(x='episode', y='rolling', data=data_random_reward, label='random_rolling')
#ax = sns.lineplot(x='episode', y='reward', data=data_first_reward, label='first')
#ax = sns.lineplot(x='episode', y='reward', data=data_second_reward, label='second')
```

# EPISODE LENGTH PLOT

```python
fig, ax = plt.subplots()
ax = sns.lineplot(x='episode', y='length', data=data_random_length, label='random')
ax = sns.lineplot(x='episode', y='rolling', data=data_random_length, label='random_rolling')
#ax = sns.lineplot(x='episode', y='length', data=data_first_length, label='first')
#ax = sns.lineplot(x='episode', y='length', data=data_second_length, label='second')
```
